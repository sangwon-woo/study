# Axes Customizing

```python
fig.tight_layout(ax.set_title, ax.set_xlabel, ax.set_ylabel)
```

````python
fig, ax = plt.subplots(figsize=(7, 7), facecolor='linen')
ax.set_title('Title~!', fontsize=20)
ax.set_xlabel('X Label~!', fontsize=15)
ax.set_ylabel('Y Label~!', fontsize=15)
````

````python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X label ' + str(i) for i in range(4)]
ylabel_list = ['Y label ' + str(i) for i in range(4)]

fig, axes = plt.subplots(2, 2, figsize=(10, 10), facecolor='linen')
for ax_idx, ax in enumerate(axes.flat):
    _ = ax.set_title(title_list[ax_idx], fontsize=30)
    _ = ax.set_xlabel(xlabel_list[ax_idx], fontsize=20)
    _ = ax.set_ylabel(ylabel_list[ax_idx], fontsize=20)

fig.tight_layout(pad=3)
````
